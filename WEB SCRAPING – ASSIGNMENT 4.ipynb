1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url 
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos 

You need to find following details: A)
Rank 
B) Name 
C) Artist 
D) Upload date 
E) Views 

```python
#Importing Neccessary libraries
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
import requests
```

```python
#Opening the Given page on chrome browser
driver=webdriver.Chrome()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
#B)Name
name=[]
try:
    name_tag=driver.find_elements(By.XPATH,'//table[@ class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
    for i in name_tag:
        name.append(i.text)
except NoSuchElementException:
    name.append("-")
    
```

```python
#c)Artist
artist=[]
try:
    artist_tag=driver.find_elements(By.XPATH,'//table[@ class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
    for i in artist_tag:
        artist.append(i.text)
except NoSuchElementException:
    artist.append("-")
    
```

```python
#A)Rank
rank=[]
for i in range(1,31):
    rank.append(i)
    
```

```python
#c)upload_date
upload_date=[]
try:
    upload_tag=driver.find_elements(By.XPATH,'//table[@ class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
    for i in upload_tag:
        upload_date.append(i.text)
except NoSuchElementException:
    upload_date.append("-")
    
```

```python
#c)views
views=[]
try:
    views_tag=driver.find_elements(By.XPATH,'//table[@ class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
    for i in views_tag:
        views.append(i.text)
except NoSuchElementException:
    views.append("-")
```

```python
df=pd.DataFrame(
    {
    "RANK":rank,
    "NAME":name,
    "ARTIST":artist,
    "UPLOAD DATE":upload_date,
    "VIEWS":views
})
```

```python
df
```

```python
#clicking on trending
trending=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
trending.click()
```

```python
#Extracting the url
url=[]
url_tag=driver.find_elements(By.XPATH,"//h2[@class='h3 lh-condensed']//a")
for i in url_tag:
    url.append(i.get_attribute('href'))
```

```python
#A)Repository title
title=[]
title_tag=driver.find_elements(By.XPATH,"//h2[@class='h3 lh-condensed']")
for i in title_tag:
    title.append(i.text)
    
```

```python
desc=[]
contr=[]
lang1=[]
lang=[]
for i in url:
    driver.get(i)
    time.sleep(5)
    try:
        desc_tag = driver.find_element(By.XPATH,'//p[@ class="f4 mb-3 "]')
        desc.append(desc_tag.text)
    except NoSuchElementException:
        desc.append('-')
        
    try:
        contr_tag = driver.find_element(By.XPATH,'//a[@ class="Link--inTextBlock Link"]')
        contr.append(contr_tag.text.replace("+"," "))
    except NoSuchElementException:
        contr.append('-')
    
    try:
        for i in driver.find_elements(By.XPATH,"//ul[@class= 'list-style-none']//li//span[1]"):
            lang.append(i.text)
        lang1.append(lang)
    except NoSuchElementException:
        lang1.append('-')
        
```

```python
len(title),len(desc),len(contr),len(url),len(lang1)
```

```python
#Creating dataframe
df=pd.DataFrame({
    "TITLE":title,
    "DESCRIPTION":desc,
    "CONTRIBUTORS":contr,
    "LANGUAGE":lang1
})
```
