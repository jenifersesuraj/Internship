1) Write a python program to display IMDB’s Top rated 100 Indian movies’ data
https://www.imdb.com/list/ls056092300/ 
(i.e. name, rating, year ofrelease) and make data frame.


```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get("https://www.imdb.com/list/ls056092300/")
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
soup
```

```python
movie= soup.find_all("div", class_="lister-item-content")
```

```python
names=[]
ratings = []
years = []
for i in movie:
    n = i.h3.a.text
    names.append(n)
    r = i.find("div", class_="ipl-rating-star").text.strip("\n")
    ratings.append(r)
    y = i.find("span", class_="lister-item-year").text.strip("()")
    years.append(y)
```

```python
df = pd.DataFrame({
    "Name": names,
    "Rating": ratings,
    "Year of Release": years
})
```

```python
df
```

8) Please visit https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloadedarticles/ and scrap-
 a) Paper title
 b) date
 c) Author

```python
import requests
from bs4 import BeautifulSoup
```

```python
link= "https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/"
```

```python
response = requests.get(link)
response
```

```python
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
container= soup.find_all("div", class_="article-listing")
```

```python
container
```

```python
titles = []
dates = []
authors = []
paptit= soup.find_all("h2", class_="h5 article-title")
for i in paptit:
    n=(i.text.strip())
    titles.append(n)
```

```python
for paper in container:
    date = paper.find("p", class_="article-date").text.strip()
    dates.append(date)
    author =paper.find("p", class_="article-authors").text.strip()
    authors.append(author)
```

```python
import pandas as pd
df=pd.DataFrame({
    
    "Title":titles,
    "Date":dates,
    "Authors":authors
})
df
```
