```python
#getting the city name from user
city_name = input('Enter City name : ')
#navigating to the search_bar
search_bar = driver.find_element(By.XPATH,"/html/body/div[1]/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/form/input")
search_bar.send_keys(city_name)
time.sleep(5)
```

```python
#Extracting the url using the regex finding the latitude and longitude
import re
try:
    url_str = driver.current_url
    print("URL Extracted: ", url_str)
    latitude_longitude = re.findall(r'@(.*)data',url_str)
    if len(latitude_longitude):
        lat_lng_list = latitude_longitude[0].split(",")
        if len(lat_lng_list)>=2:
            latitude = lat_lng_list[0]
            longitude = lat_lng_list[1]
        print("Latitude = {}, Longitude = {}".format(latitude, longitude))
except Exception as e:
        print("Error: ", str(e))
```

2. In the above question, now scrape the following details of each product listed in first 3 pages of your search 
results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then 
scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 

```python
# Extracting Product url
urls = []          
for i in range(0,3):      
    page_url = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-no-outline"]')
    for i in page_url:
        try:
            urls.append(i.get_attribute("href"))
            next_btn = driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[65]/div/div/span/a[3]")
            time.sleep(5)
        except NoSuchElementException:
            urls.append("-")
```

```python
#Extracting name
brand_name=[]
for url in urls:
    if url=='-':
        brand_name.append("-")
        continue
    driver.get(url)
    time.sleep(3)
    try:
        brand_ntag=driver.find_element(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        brand_name.append(brand_ntag.text)
    except NoSuchElementException:
        brand_name.append('-')
```

```python
#Extracting price
price=[]
for url in urls:
    if url=='-':
        price.append("-")
        continue
    driver.get(url)
    time.sleep(3)
    try:
        price_tag=driver.find_element(By.XPATH,'//span[@ class="a-price-whole"]')
        price.append(price_tag.text)
    except NoSuchElementException:
        price.append('-')
```

```python
#Extracting availability
avail=[]
for url in urls:
    if url=='-':
        avail.append("-")
        continue
    driver.get(url)
    time.sleep(3)
    try:
        avail_tag=driver.find_element(By.XPATH,'//span[@ class="a-size-medium a-color-success"]')
        avail.append(avail_tag.text)
    except NoSuchElementException:
        avail.append('-')
        
```

```python
#Extracting expected delivery
exp_del=[]
for url in urls:
    if url=='-':
        exp_del.append("-")
        continue
    driver.get(url)
    time.sleep(3)
    try:
        exp_tag=driver.find_element(By.XPATH,'//span[@ class="a-text-bold"][1]')
        exp_del.append(exp_tag.text)
    except NoSuchElementException:
        exp_del.append('-')
    
```

```python
#Extracting return
ret=[]
for url in urls:
    if url=='-':
        ret.append("-")
        continue
    driver.get(url)
    time.sleep(3)
    try:
        ret_tag=driver.find_element(By.XPATH,"//span[@class='a-size-small a-color-link a-text-normal'][normalize-space()='7 days Replacement']")
        ret.append(ret_tag.text)
    except NoSuchElementException:
        ret.append('-')
```

```python
#checking the length
len(brand_name),len(urls),len(price),len(avail),len(exp_del),len(ret)
```

```python
#creating dataframe
df=pd.DataFrame({
    "PRODUCT TITLE":brand_name,
    "PRODUCT URL":urls,
    "PRICE":price,
    "AVAILABILITY":avail,
    "EXPECTED DELIVERY":exp_del,
    "RETURN":ret
})
```
