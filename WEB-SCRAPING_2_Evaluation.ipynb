Q7: Write a python program to display list of respected former Prime Ministers of India (i.e. Name, Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1 scrap the mentioned data and make the DataFrame

```python
#Importing Neccessary libraries
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
#Opening the Given page on chrome browser
driver=webdriver.Chrome()
driver.get("https://www.jagranjosh.com/general-knowledge/list-of%02all-prime-ministers-of-india-1473165149-1")
```

```python
#Extracting information from the table with help of xpath and tagname
data = []
table1 = driver.find_element(By.XPATH,'//div[@class="TableData"]')
rows = table1.find_elements(By.TAG_NAME,'tr')
rows
```

```python
#extracting information from each column and  appending the information to the corresponding named empty list
names=[]
born_deads=[]
terms_of_office=[]
remarks=[]
for row in rows:
    cols = row.find_elements(By.TAG_NAME,'td')
    if len(cols) == 5:
        name= cols[1].text
        names.append(name)
        born_dead= cols[2].text
        born_deads.append(born_dead)
        term= cols[3].text
        terms_of_office.append(term)
        remark=cols[4].text
        remarks.append(remark)
```

```python
#creating the dataframe
df=pd.DataFrame({
    "NAME":names,
    "BORN DEAD":born_deads,
    "TERM OF OFFICE":terms_of_office,
    "REMARKS":remarks
    
})
```

Q3: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link: https://www.flipkart.com/apple-iphone-11-black-64-gb/productreviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART As shown in the above page you have to scrape the tick marked attributes. These are:

Rating Review summary Full review You have to scrape this data for first 100review

```python
# Let's now import all the required libraries.
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
#Opening the given page through chrome browser
driver=webdriver.Chrome()
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART")
```

```python
#function for full review
def full_review_fn():
    full_review=[]
    driver.refresh()
    rev_tag=driver.find_elements(By.XPATH,"//div[@class='t-ZTKy']/div")
    time.sleep(3)
    for i in rev_tag:
        try:
            full_review.append(i.text.replace("\n","  New Line: "))
            driver.execute_script("window.scrollBy(0,document.body.scrollHeight0)")
        except:
            full_review.append("--")
    return full_review
```

```python
#function for review summary
def review_summary_fn():
    review_sum=[]
    driver.refresh()
    sum_tag=driver.find_elements(By.XPATH,"//p[@class='_2-N8zT']")
    time.sleep(3)
    for i in sum_tag:
        try:
            review_sum.append(i.text)
            driver.execute_script("window.scrollBy(0,document.body.scrollHeight0)")
        except:
            review_sum.append("--")
    return review_sum
            
```

```python
#function for rating
def rating_fn():
    rating=[]
    driver.refresh()
    rating_tag=driver.find_elements(By.XPATH,"//div[@class='_3LWZlK _1BLPMq']")
    time.sleep(3)
    for i in rating_tag:
        try:
            rating.append(i.text)
            driver.execute_script("window.scrollBy(0,document.body.scrollHeight0)")
        except:
            rating.append("--")
    return rating
```

```python
#scraping the information from the function
rating=[]
review_sum=[]
full_review=[]
length=len(rating)
while(length<=100):
    driver.refresh()
    rating.extend(rating_fn())
    review_sum.extend(review_summary_fn())
    full_review.extend(full_review_fn())
    time.sleep(5)
    length=len(rating)
    next_btn=driver.find_element(By.XPATH,"//a[@class='_1LKTO3']/span")
    next_btn.click()
```

```python
len(full_review),len(rating),len(review_sum)
```

```python
#creating the dataframe and adding the column
import pandas as pd
df=pd.DataFrame()
df["RATING"]=rating[:100]
df["REVIEW_SUMMARY"]=review_sum[:100]
df["FULL_REVIEW"]=full_review[:100]

```

```python
#displaying the dataframe
df
```
